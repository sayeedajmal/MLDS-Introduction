# Two Options for Subplots

```python
#Option 1
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

 #plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.random(1000));
```

```python
#Option 2
fig, ax=plt.subplots(nrows=2, ncols=2,figsize=(10,5))

#Plot to each Different Axis
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

# Ploting From Pandas DataFrame

```python
#Make DF
car_sales=pd.read_csv("car-sales.csv");
car_sales
```

```python
#Remove $ Sign from Price
car_sales["Price"] = car_sales["Price"].str.replace('[\$,]', '', regex=True).str.split('.').str[0].astype(int)
car_sales
```

```python
#Add Date Range Column
car_sales["Sale Date"]=pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
#Adding Total Sales by cumsum() and the Price convert from str to int
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
#Plot the Total Sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)",y="Price", kind="scatter")
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head()
```

### Which one Shoud I use? (Pyplot or Matplotlib OO Method)
* When Plotting something Quickly, okey to use pyplt method
* When plotting more advance, use the OO method

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
len(over_50)
```

```python
#Pyplot Method
over_50.plot(kind="scatter", x='age', y='chol', c='target');
```

```python
# OO Method
fig, ax=plt.subplots(figsize=(10,5))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
```

```python
## OO Method From Scratch
fig, ax=plt.subplots(figsize=(10,5))

#Plot the Data
scatter=ax.scatter(x=over_50["age"], y=over_50["chol"],c=over_50["target"]);

#Customize
ax.set(title="Heart Disease", xlabel="Age", ylabel="cholestrol");

#Add Legend
ax.legend(*scatter.legend_elements(), title="Target");
#Add A Horizontal Line
ax.axhline(over_50["chol"].mean(),linestyle='--' );
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach

fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1, figsize=(10,10),sharex=True);
#AX1
scatter=ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter");
ax0.set(title="Heart Disease and Colestrol Levels", ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(), title="Target",);
ax0.axhline(y=over_50["chol"].mean(),linestyle="--", label="Average");

#AX2
scatter=ax1.scatter(x=over_50["age"], y=over_50["thalach"],c=over_50["target"],cmap="winter");
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--", label="Average");


#TITILE TO FIGURE
fig.suptitle("HEART DISEASE ANALYSIS", fontsize=16, fontweight='bold');
```
